```python
features = dataset[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
label = dataset['diagnosis']
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(features)
scaled = scalar.transform(features)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled, label, test_size=0.2, stratify=label)
```

```python
from sklearn import svm
from sklearn.metrics import accuracy_score

clf = svm.SVC(kernel='linear')
clf.fit(x_train, y_train)
y_pred_svm = clf.predict(x_test)

score_svm = accuracy_score(y_test, y_pred_svm)

score_svm
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

clf_nn = MLPClassifier(random_state=1, max_iter=900).fit(x_train, y_train)
clf_nn.fit(x_train, y_train)
y_pred_mlpc = clf.predict(x_test)
score_mlpc = accuracy_score(y_test, y_pred_mlpc)

score_mlpc
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf_rf=RandomForestClassifier(n_estimators=100)
clf_rf.fit(x_train, y_train)
y_pred_rf = clf.predict(x_test)
score_rf = accuracy_score(y_test, y_pred_rf)

score_rf
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=15)
principalComponents = pca.fit_transform(scaled)
x_train, x_test, y_train, y_test = train_test_split(principalComponents, label, test_size=0.2, stratify=label)

clf = svm.SVC(kernel='linear')
clf.fit(x_train, y_train)
y_pred_svm = clf.predict(x_test)

score_svm_pca = accuracy_score(y_test, y_pred_svm)

score_svm_pca
```

```python
clf_nn = MLPClassifier(random_state=1, max_iter=900).fit(x_train, y_train)
clf_nn.fit(x_train, y_train)
y_pred_mlpc = clf.predict(x_test)
score_mlpc_pca = accuracy_score(y_test, y_pred_mlpc)

score_mlpc_pca
```

```python
clf_rf=RandomForestClassifier(n_estimators=100)
clf_rf.fit(x_train, y_train)
y_pred_rf = clf.predict(x_test)
score_rf_pca = accuracy_score(y_test, y_pred_rf)

score_rf_pca
```

```python
import matplotlib.pyplot as plt


name = ['SVM(pre) ', 'MPLC(pre)', 'RF(pre)','SVM(post)', 'MPLC(post)', 'RF(post)']
scores = [score_svm, score_mlpc, score_rf, score_svm_pca, score_mlpc_pca, score_rf_pca]

x_pos = np.arange(len(name))


plt.xticks(x_pos, name)
plt.bar(x_pos, scores)
```
